```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as mtp
#Importing the dataset
df=pd.read_csv('Mall_Customers.csv')
df
```

```python
# Extracting features
x=df.iloc[:,[3,4]].values
x
```

```python
#Finding the optimal number of clusters using the dendrogram
import scipy.cluster.hierarchy as shc
dendro=shc.dendrogram(shc.linkage(x,method="complete"))
mtp.title("Dendrogram Plot")
mtp.ylabel("Euclidean Distances")
mtp.xlabel("Customers")
mtp.show()
```

```python
#training the hierarchical model on dataset
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='complete')

yPred=hc.fit_predict(x)
yPred
```

```python
#visulaizing the clusters
mtp.scatter(x[yPred==0,0],x[yPred==0,1],s=100,c='blue',label='Cluster 1')
mtp.scatter(x[yPred==1,0],x[yPred==1,1],s=100,c='green',label='Cluster 2')
mtp.scatter(x[yPred==2,0],x[yPred==2,1],s=100,c='red',label='Cluster 2')
mtp.scatter(x[yPred==3,0],x[yPred==3,1],s=100,c='cyan',label='Cluster 2')
mtp.scatter(x[yPred==4,0],x[yPred==4,1],s=100,c='magenta',label='Cluster 2')

mtp.title('Clusters of Customers')
mtp.xlabel('Annual Income')
mtp.ylabel('Spending Score')
mtp.show()
```
